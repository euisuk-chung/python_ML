```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
import time

def get_post_links_and_titles(username):
    """
    Fetches all post links and titles for a given Velog username.
    
    Args:
    username (str): The Velog username.
    
    Returns:
    list: A list of dictionaries containing the title and URL of the user's posts.
    """
    url = f'https://velog.io/@{username}'
    
    # Set up the Selenium WebDriver (Assuming using ChromeDriver)
    chrome_driver_path = 'D:/webdriver/chromedriver-win64/chromedriver.exe'  # ChromeDriverì˜ ì‹¤ì œ ê²½ë¡œë¡œ ë³€ê²½í•˜ì„¸ìš”
    service = Service(chrome_driver_path)
    options = webdriver.ChromeOptions()
    options.add_argument("--headless")  # ë¸Œë¼ìš°ì € ì°½ì„ ì—´ì§€ ì•ŠìŒ
    driver = webdriver.Chrome(service=service, options=options)
    
    driver.get(url)
    
    # Scroll down to the bottom of the page to load all content
    last_height = driver.execute_script("return document.body.scrollHeight")
    
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)  # Wait for new content to load
        
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
    
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    # HTML ì¶œë ¥ ì €ì¥
    html_output = soup.prettify()
    with open('html_output.txt', 'w', encoding='utf-8') as file:
        file.write(html_output)
    
    posts = []
    for div in soup.find_all('div', class_='FlatPostCard_block__a1qM7'):
        a_tag = div.find('a', class_='VLink_block__Uwj4P')
        title_tag = div.find('h2')  # h2 ëŒ€ì‹  h4 ì‚¬ìš©
        if a_tag:
            href = a_tag.get('href', '')
            title = title_tag.get_text(strip=True) if title_tag else 'No Title'
            print(title)
            if href:
                posts.append({
                    'title': title,
                    'url': href
                })
    
    driver.quit()
    return posts

if __name__ == "__main__":
    username = 'euisuk-chung'  # ì‚¬ìš©ìì˜ Velog ID
    posts = get_post_links_and_titles(username)
    print(posts)  # Print the extracted posts for inspection
```
