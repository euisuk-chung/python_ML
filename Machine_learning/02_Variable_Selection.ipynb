### 1. ëª¨ë“ˆ ë¶ˆëŸ¬ì˜¤ê¸°

```python
# import ë¶ˆëŸ¬ì˜¬ íŒ¨í‚¤ì§€ëª… as ê·¸ íŒ¨í‚¤ì§€ë¥¼ íŒŒì´ì¬ì—ì„œ ì‚¬ìš©í•  ì´ë¦„
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import matplotlib.pyplot as plt

# í•œê¸€ í°íŠ¸ ì„¤ì •
plt.rc('font', family='Malgun Gothic')

# ë…¸íŠ¸ë¶ìƒì—ì„œ ì´ì˜ê²Œ ì¶œë ¥í•˜ê¸° ìœ„í•´
from IPython.display import display, HTML
```

### 2. ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°

```python
# Load data
data = pd.read_csv('./Data/BostonHousing.csv')
data.head()
```

### 3. ë°ì´í„° ì¤€ë¹„ 
#### Train set / Test set ë‚˜ëˆ„ê¸°

```python
X_cols = data.columns.drop('MEDV')
X = data[X_cols]
y = data['MEDV']
```

```python
# ëª¨ë¸ ìƒì„±
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

### 4.2. ì„ í˜•íšŒê·€ ëª¨ë¸ í‰ê°€í•˜ê¸°

- P>[t] (p-value):0.05(ìœ ì˜ìˆ˜ì¤€) ì´í•˜ì¼ ë•Œ ë³€ìˆ˜ê°€ ìœ ì˜ë¯¸

```python
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('R^2 Score:', r2_score(y_test, y_pred))
print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))
```

### ì „ì§„ ì„ íƒ (Forward Selection)

ì „ì§„ ì„ íƒì€ ë³€ìˆ˜ê°€ ì—†ëŠ” ìƒíƒœì—ì„œ ì‹œì‘í•˜ì—¬ ë‹¨ê³„ë³„ë¡œ ë³€ìˆ˜ë¥¼ ì¶”ê°€í•˜ëŠ” ë°©ì‹ì…ë‹ˆë‹¤. ê° ë‹¨ê³„ë§ˆë‹¤ ëª¨ë¸ì˜ ì„±ëŠ¥ì„ ê°€ì¥ ë§ì´ ê°œì„ í•˜ëŠ” ë³€ìˆ˜ë¥¼ ì„ íƒí•˜ì—¬ ì¶”ê°€í•©ë‹ˆë‹¤. 

#### ê³¼ì •
1. **ì´ˆê¸° ìƒíƒœ**: ëª¨ë¸ì— ì•„ë¬´ëŸ° ë³€ìˆ˜ê°€ í¬í•¨ë˜ì–´ ìˆì§€ ì•ŠìŠµë‹ˆë‹¤.
2. **1ë‹¨ê³„**:
    - ëª¨ë“  ë³€ìˆ˜ ì¤‘ì—ì„œ í•˜ë‚˜ì”© ì¶”ê°€í•˜ì—¬ ëª¨ë¸ì„ í‰ê°€í•©ë‹ˆë‹¤.
    - `CRIM`, `ZN`, `INDUS`, `NOX`, `RM`, `AGE`, `DIS`, `TAX`, `PTRATIO`, `B`, `LSTAT` ê°ê°ì„ ëª¨ë¸ì— ì¶”ê°€í•œ í›„ ëª¨ë¸ì˜ ì„±ëŠ¥ì„ í‰ê°€í•©ë‹ˆë‹¤.
    - ëª¨ë¸ ì„±ëŠ¥ì´ ê°€ì¥ ë§ì´ ê°œì„ ë˜ëŠ” ë³€ìˆ˜ë¥¼ ì„ íƒí•©ë‹ˆë‹¤. ì˜ˆë¥¼ ë“¤ì–´, `RM`ì´ ì„ íƒë˜ì—ˆë‹¤ê³  ê°€ì •í•©ë‹ˆë‹¤.
3. **2ë‹¨ê³„**:
    - í˜„ì¬ í¬í•¨ëœ ë³€ìˆ˜ëŠ” `RM`ì…ë‹ˆë‹¤.
    - ë‚˜ë¨¸ì§€ ë³€ìˆ˜ ì¤‘ì—ì„œ í•˜ë‚˜ì”© ì¶”ê°€í•˜ì—¬ ëª¨ë¸ì„ í‰ê°€í•©ë‹ˆë‹¤.
    - `CRIM`, `ZN`, `INDUS`, `NOX`, `AGE`, `DIS`, `TAX`, `PTRATIO`, `B`, `LSTAT` ê°ê°ì„ ì¶”ê°€í•˜ì—¬ ëª¨ë¸ì˜ ì„±ëŠ¥ì„ í‰ê°€í•©ë‹ˆë‹¤.
    - ëª¨ë¸ ì„±ëŠ¥ì´ ê°€ì¥ ë§ì´ ê°œì„ ë˜ëŠ” ë³€ìˆ˜ë¥¼ ì„ íƒí•©ë‹ˆë‹¤. ì˜ˆë¥¼ ë“¤ì–´, `LSTAT`ì´ ì„ íƒë˜ì—ˆë‹¤ê³  ê°€ì •í•©ë‹ˆë‹¤.
4. **3ë‹¨ê³„**:
    - í˜„ì¬ í¬í•¨ëœ ë³€ìˆ˜ëŠ” `RM`, `LSTAT`ì…ë‹ˆë‹¤.
    - ë‚˜ë¨¸ì§€ ë³€ìˆ˜ ì¤‘ì—ì„œ í•˜ë‚˜ì”© ì¶”ê°€í•˜ì—¬ ëª¨ë¸ì„ í‰ê°€í•©ë‹ˆë‹¤.
    - `CRIM`, `ZN`, `INDUS`, `NOX`, `AGE`, `DIS`, `TAX`, `PTRATIO`, `B` ê°ê°ì„ ì¶”ê°€í•˜ì—¬ ëª¨ë¸ì˜ ì„±ëŠ¥ì„ í‰ê°€í•©ë‹ˆë‹¤.
    - ì´ ê³¼ì •ì„ ë” ì´ìƒ ìœ ì˜ë¯¸í•œ ì„±ëŠ¥ í–¥ìƒì´ ì—†ì„ ë•Œê¹Œì§€ ë°˜ë³µí•©ë‹ˆë‹¤.

```python
def forward_selection(X_train, y_train):
    model = LinearRegression()
    sfs = SequentialFeatureSelector(model, n_features_to_select='auto', direction='forward', scoring='r2').fit(X_train, y_train)
    selected_features = X_train.columns[sfs.get_support()]
    return selected_features
```

```python
selected_features_forward = forward_selection(X_train, y_train)
print(f'ì „ì§„ ì„ íƒìœ¼ë¡œ ì„ íƒëœ íŠ¹ì„±: {selected_features_forward}')
```

2. **1ë‹¨ê³„**:
    - ëª¨ë“  ë³€ìˆ˜ê°€ í¬í•¨ëœ ìƒíƒœì—ì„œ ëª¨ë¸ì„ í‰ê°€í•©ë‹ˆë‹¤.
    - ê° ë³€ìˆ˜ì˜ ì¤‘ìš”ë„ë¥¼ í‰ê°€í•˜ì—¬ ëª¨ë¸ ì„±ëŠ¥ì— ê°€ì¥ ì ì€ ì˜í–¥ì„ ë¯¸ì¹˜ëŠ” ë³€ìˆ˜ë¥¼ ì‹ë³„í•©ë‹ˆë‹¤.
    - ì˜ˆë¥¼ ë“¤ì–´, `AGE`ê°€ ê°€ì¥ ì˜í–¥ì´ ì ì€ ë³€ìˆ˜ë¡œ íŒë‹¨ë˜ì—ˆë‹¤ë©´ ì´ë¥¼ ì œê±°í•©ë‹ˆë‹¤.

    - í˜„ì¬ í¬í•¨ëœ ë³€ìˆ˜ëŠ” `CRIM`, `ZN`, `INDUS`, `NOX`, `RM`, `DIS`, `TAX`, `PTRATIO`, `B`, `LSTAT`ì…ë‹ˆë‹¤.
    - ë‚˜ë¨¸ì§€ ë³€ìˆ˜ ì¤‘ì—ì„œ ëª¨ë¸ ì„±ëŠ¥ì— ê°€ì¥ ì ì€ ì˜í–¥ì„ ë¯¸ì¹˜ëŠ” ë³€ìˆ˜ë¥¼ ì‹ë³„í•©ë‹ˆë‹¤.
    - `ZN`ì´ ê°€ì¥ ì˜í–¥ì´ ì ì€ ë³€ìˆ˜ë¡œ íŒë‹¨ë˜ì—ˆë‹¤ë©´ ì´ë¥¼ ì œê±°í•©ë‹ˆë‹¤.
4. **3ë‹¨ê³„**:
    - í˜„ì¬ í¬í•¨ëœ ë³€ìˆ˜ëŠ” `CRIM`, `INDUS`, `NOX`, `RM`, `DIS`, `TAX`, `PTRATIO`, `B`, `LSTAT`ì…ë‹ˆë‹¤.
    - ì´ ê³¼ì •ì„ ë” ì´ìƒ ë³€ìˆ˜ë¥¼ ì œê±°í•´ë„ ëª¨ë¸ ì„±ëŠ¥ì´ ìœ ì˜ë¯¸í•˜ê²Œ ë–¨ì–´ì§€ì§€ ì•Šì„ ë•Œê¹Œì§€ ë°˜ë³µí•©ë‹ˆë‹¤.

```python
def backward_elimination(X_train, y_train):
    model = LinearRegression()
    sfs = SequentialFeatureSelector(model, n_features_to_select='auto', direction='backward', scoring='r2').fit(X_train, y_train)
    selected_features = X_train.columns[sfs.get_support()]
    return selected_features
```

```python
selected_features_backward = backward_elimination(X_train, y_train)
print(f'í›„ì§„ ì œê±°ë¡œ ì„ íƒëœ íŠ¹ì„±: {selected_features_backward}')
```

### ë‹¨ê³„ì  ì„ íƒ (Stepwise Selection)
