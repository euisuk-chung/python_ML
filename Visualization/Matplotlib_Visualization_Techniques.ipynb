## 상관관계 (Correlation)

상관관계 플롯은 두 개 이상의 변수 간의 관계를 시각화하는 데 사용됩니다.


### 1. 산점도 (Scatter plot)

산점도는 두 변수 간의 관계를 연구하기 위해 사용되는 기본적이고 고전적인 플롯입니다.


````python
# 산점도
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title('산점도 예제')
plt.xlabel('꽃받침 길이')
plt.ylabel('꽃받침 너비')
plt.show()
````

### 2. 원형 강조된 버블 차트 (Bubble plot with Encircling)

특정 경계 내에 있는 점 그룹을 강조하여 표시하고자 할 때 유용합니다.


````python
# 버블 차트
x = np.random.rand(30)
y = np.random.rand(30)
size = 1000 * np.random.rand(30)

plt.scatter(x, y, s=size, alpha=0.5)
plt.title('원형 강조된 버블 차트')
plt.xlabel('X 축')
plt.ylabel('Y 축')

highlight = (x > 0.7) & (y > 0.7)
plt.scatter(x[highlight], y[highlight], s=size[highlight], facecolors='none', edgecolors='r')

plt.show()
````

### 3. 회귀선이 포함된 산점도 (Scatter plot with line of best fit)

두 변수 간의 변화 관계를 이해하고자 할 때 회귀선이 있는 산점도가 유용합니다.

### 4. 지터링을 사용한 스트립플롯 (Jittering with stripplot)

여러 데이터 포인트가 동일한 X 및 Y 값을 갖는 경우 포인트들이 겹쳐져서 보이지 않을 수 있습니다. 이를 피하기 위해 포인트를 약간씩 지터링하면 시각적으로 볼 수 있습니다.

### 6. 주변 히스토그램 (Marginal Histogram)

주변 히스토그램은 X 및 Y 축 변수와 함께 히스토그램을 갖습니다. 이는 X와 Y의 관계를 시각화하면서 X와 Y 각각의 단변량 분포를 시각화하는 데 사용됩니다.


````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# penguin 데이터셋 로드 (이 예제에서는 seaborn의 penguins 데이터셋을 사용하겠습니다)
penguin = sns.load_dataset("penguins")

# Create Fig and gridspec
fig = plt.figure(figsize=(16, 10), dpi= 80)
grid = plt.GridSpec(4, 4, hspace=0.5, wspace=0.2)

# Define the axes
ax_main = fig.add_subplot(grid[:-1, :-1])
ax_right = fig.add_subplot(grid[:-1, -1], xticklabels=[], yticklabels=[])
ax_bottom = fig.add_subplot(grid[-1, 0:-1], xticklabels=[], yticklabels=[])

# Scatterplot on main ax
sns.scatterplot(ax=ax_main, x='bill_length_mm', y='bill_depth_mm', hue='body_mass_g', palette='viridis', data=penguin, s=150, alpha=0.7)
ax_main.get_legend().remove()

# Histogram in the bottom
sns.histplot(ax=ax_bottom, x=penguin['bill_length_mm'], bins=50, kde=True, color='skyblue')

# Histogram on the right
sns.histplot(ax=ax_right, y=penguin['bill_depth_mm'], bins=50, kde=True, color='skyblue')
ax_bottom.invert_yaxis()

# Decorations
ax_main.set(title='Scatterplot with Histograms \n bill_length_mm vs bill_depth_mm', xlabel='length', ylabel='depth')
ax_main.title.set_fontsize(20)

for item in ([ax_main.xaxis.label, ax_main.yaxis.label] + ax_main.get_xticklabels() + ax_main.get_yticklabels()):
    item.set_fontsize(14)

xlabels = ax_main.get_xticks().tolist()
ax_main.set_xticklabels(xlabels)
plt.show()
````

````python
# 주변 히스토그램 using jointplot
plt.figure(figsize=(10, 6))
sns.jointplot(x='bill_length_mm', y='bill_depth_mm', palette='viridis', data=penguin, kind='scatter', marginal_kws=dict(bins=15, fill=True))
plt.show()
````

### 7. 주변 상자 그림 (Marginal Boxplot)

주변 상자 그림은 주변 히스토그램과 유사한 목적을 가지고 있습니다. 그러나 상자 그림은 X와 Y의 중앙값, 25번째 및 75번째 백분위수를 정확하게 파악하는 데 도움이 됩니다.


````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# penguin 데이터셋 로드 (이 예제에서는 seaborn의 penguins 데이터셋을 사용하겠습니다)
penguin = sns.load_dataset("penguins")

# Create Fig and gridspec
fig = plt.figure(figsize=(16, 10), dpi=80)
grid = plt.GridSpec(4, 4, hspace=0.5, wspace=0.2)

# Define the axes
ax_main = fig.add_subplot(grid[:-1, :-1])
ax_right = fig.add_subplot(grid[:-1, -1], xticklabels=[], yticklabels=[])
ax_bottom = fig.add_subplot(grid[-1, 0:-1], xticklabels=[], yticklabels=[])

# Scatterplot on main ax
sns.scatterplot(ax=ax_main, x='bill_length_mm', y='bill_depth_mm', hue='body_mass_g', palette='viridis', data=penguin, s=150, alpha=0.7)
ax_main.get_legend().remove()

# Boxplot in the bottom
sns.boxplot(ax=ax_bottom, x=penguin['bill_length_mm'], color='red')
ax_bottom.set(yticks=[], ylabel='')

# Boxplot on the right
sns.boxplot(ax=ax_right, y=penguin['bill_depth_mm'], color='red')
ax_right.set(xticks=[], xlabel='')

# Decorations
ax_main.set(title='Scatterplot with Marginal Boxplots \n bill_length_mm vs bill_depth_mm', xlabel='length', ylabel='depth')
ax_main.title.set_fontsize(20)

for item in ([ax_main.xaxis.label, ax_main.yaxis.label] + ax_main.get_xticklabels() + ax_main.get_yticklabels()):
    item.set_fontsize(14)

plt.show()
````

### 10. 발산 막대 그래프 (Diverging Bars)

단일 메트릭을 기반으로 항목이 어떻게 변하는지 보고 이러한 변동의 순서와 양을 시각화하려면 발산 막대 그래프가 유용합니다.


````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl

# 폰트 설정
plt.rcParams["font.family"] = 'NanumGothic'
mpl.rcParams['axes.unicode_minus'] = False  # 마이너스 폰트 깨짐 방지

# Prepare Data
df = pd.read_csv("https://github.com/selva86/datasets/raw/master/mtcars.csv")
x = df.loc[:, ['mpg']]
df['mpg_z'] = (x - x.mean())/x.std()
df['colors'] = ['red' if x < 0 else 'green' for x in df['mpg_z']]
df.sort_values('mpg_z', inplace=True)
df.reset_index(inplace=True)

plt.figure(figsize=(14, 10), dpi= 80)
plt.hlines(y=df.index, xmin=0, xmax=df.mpg_z, color=df.colors, alpha=0.4, linewidth=5)
plt.gca().set(ylabel='$Model$', xlabel='$Mileage$')
plt.yticks(df.index, df.cars, fontsize=12)
plt.title('발산 막대 그래프 (Diverging Bars)', fontdict={'size':20})
plt.grid(linestyle='--', alpha=0.5)
plt.show()
````

### 11. 발산 텍스트 그래프 (Diverging Texts)

발산 텍스트 그래프는 발산 막대 그래프와 유사하며 각 항목의 값을 시각적으로 표시하는 데 유용합니다.


````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl

# 폰트 설정
plt.rcParams["font.family"] = 'NanumGothic'
mpl.rcParams['axes.unicode_minus'] = False  # 마이너스 폰트 깨짐 방지

# 발산 텍스트 그래프
df = pd.read_csv("https://github.com/selva86/datasets/raw/master/mtcars.csv")
x = df.loc[:, ['mpg']]
df['mpg_z'] = (x - x.mean())/x.std()
df['colors'] = ['red' if x < 0 else 'green' for x in df['mpg_z']]
df.sort_values('mpg_z', inplace=True)
df.reset_index(inplace=True)

# Draw plot
plt.figure(figsize=(14,14), dpi= 80)
plt.hlines(y=df.index, xmin=0, xmax=df.mpg_z, color=df.colors)
for x, y, tex in zip(df.mpg_z, df.index, df.mpg_z):
    t = plt.text(x, y, round(tex, 2), horizontalalignment='right' if x < 0 else 'left', 
                 verticalalignment='center', fontdict={'color':'red' if x < 0 else 'green', 'size':14})

# Decorations    
plt.yticks(df.index, df.cars, fontsize=12)
plt.title('발산 텍스트 그래프 (Diverging Texts)', fontdict={'size':20})
plt.grid(linestyle='--', alpha=0.5)
plt.xlim(-2.5, 2.5)
plt.show()
````

### 15. 순서가 있는 막대 그래프 (Ordered Bar Chart)

````python
# 순서가 있는 막대 그래프
df = sns.load_dataset('mpg')
df = df[['cty', 'manufacturer']].groupby('manufacturer').apply(lambda x: x.mean())
df.sort_values('cty', inplace=True)
df.reset_index(inplace=True)

fig, ax = plt.subplots(figsize=(16, 10))
ax.barh(df.manufacturer, df.cty, color='skyblue')
ax.set_title('순서가 있는 막대 그래프 (Ordered Bar Chart)', fontdict={'size':22})
ax.set_xlabel('평균 시내 연비 (MPG)')
plt.show()
````

### 16. 롤리팝 차트 (Lollipop Chart)

롤리팝 차트는 순서가 있는 막대 그래프와 비슷한 역할을 합니다.


````python
# 롤리팝 차트
df = sns.load_dataset('mpg')
df = df[['cty', 'manufacturer']].groupby('manufacturer').apply(lambda x: x.mean())
df.sort_values('cty', inplace=True)
df.reset_index(inplace=True)

fig, ax = plt.subplots(figsize=(16, 10))
ax.stem(df.index, df.cty, linefmt='-', markerfmt='o', basefmt=' ')
ax.set_title('롤리팝 차트 (Lollipop Chart)', fontdict={'size':22})
ax.set_xlabel('평균 시내 연비 (MPG)')
ax.set_xticks(df.index)
ax.set_xticklabels(df.manufacturer.str.title(), rotation=90)
plt.show()
````

### 17. 점 플롯 (Dot Plot)

점 플롯은 항목의 순서를 전달합니다. 수평 축을 따라 정렬되어 포인트 간의 거리를 쉽게 시각화할 수 있습니다.


````python
# 점 플롯
df = sns.load_dataset('mpg')
df = df[['cty', 'manufacturer']].groupby('manufacturer').apply(lambda x: x.mean())
df.sort_values('cty', inplace=True)
df.reset_index(inplace=True)

fig, ax = plt.subplots(figsize=(16, 10))
ax.hlines(y=df.index, xmin=11, xmax=26, color='gray', alpha=0.7, linewidth=1, linestyles='dashdot')
ax.scatter(y=df.index, x=df.cty, s=75, color='firebrick', alpha=0.7)
ax.set_title('점 플롯 (Dot Plot) - 시내 연비 (MPG)', fontdict={'size':22})
ax.set_xlabel('평균 시내 연비 (MPG)')
ax.set_yticks(df.index)
ax.set_yticklabels(df.manufacturer.str.title(), fontdict={'horizontalalignment': 'right'})
ax.set_xlim(10, 27)
plt.show()
````

### 37. 자기 상관 (Autocorrelation) 및 부분 상관 (Partial Autocorrelation) 플롯

자기 상관 플롯은 시계열과 자신의 랙 간의 상관 관계를 보여줍니다.


````python
# 자기 상관 및 부분 상관 플롯
df = pd.read_csv('https://github.com/selva86/datasets/raw/master/AirPassengers.csv')

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
plot_acf(df['#Passengers'], ax=ax1, lags=50)
plot_pacf(df['#Passengers'], ax=ax2, lags=20)

ax1.spines["top"].set_alpha(.3); ax2.spines["top"].set_alpha(.3)
ax1.spines["bottom"].set_alpha(.3); ax2.spines["bottom"].set_alpha(.3)
ax1.spines["right"].set_alpha(.3); ax2.spines["right"].set_alpha(.3)
ax1.spines["left"].set_alpha(.3); ax2.spines["left"].set_alpha(.3)

ax1.tick_params(axis='both', labelsize=12)
ax2.tick_params(axis='both', labelsize=12)
plt.show()
````

### 38. 교차 상관 플롯 (Cross Correlation Plot)

교차 상관 플롯은 두 시계열의 랙 간의 상관 관계를 보여줍니다.

### 39. 시계열 분해 플롯 (Time Series Decomposition Plot)

### 42. 오류 밴드가 포함된 시계열 (Time Series with Error Bands)

여러 시간 포인트에 대해 여러 관측값이 있는 시계열 데이터셋이 있는 경우 오류 밴드가 포함된 시계열을 만들 수 있습니다.


````python
# 오류 밴드가 포함된 시계열
np.random.seed(0)
dates = pd.date_range('2023-01-01', periods=100)
values = np.random.randn(100).cumsum()
errors = np.random.randn(100) * 0.5

plt.figure(figsize=(10, 6))
plt.plot(dates, values, label='값')
plt.fill_between(dates, values - errors, values + errors, color='gray', alpha=0.2)
plt.title('오류 밴드가 포함된 시계열 (Time Series with Error Bands)')
plt.xlabel('날짜')
plt.ylabel('값')
plt.show()
````

### 43. 누적 면적 차트 (Stacked Area Chart)

누적 면적 차트는 여러 시계열의 기여도를 시각적으로 나타내어 서로 비교하기 쉽게 만듭니다.

### 48. 클러스터 플롯 (Cluster Plot)

클러스터 플롯은 같은 클러스터에 속하는 포인트를 구분합니다.


````python
# 클러스터 플롯
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/USArrests.csv')
X = df[['Murder', 'Assault', 'UrbanPop', 'Rape']]

kmeans = KMeans(n_clusters=5, random_state=0).fit(X)
df['cluster'] = kmeans.labels_

plt.figure(figsize=(10, 6))
plt.scatter(df['Murder'], df['Assault'], c=df['cluster'], cmap='tab10')
plt.title('클러스터 플롯 (Cluster Plot)')
plt.xlabel('살인율')
plt.ylabel('폭행률')
plt.show()
````
